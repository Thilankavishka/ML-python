```python
import numpy
speed = [32,111,138,28,59,77,97]
x = numpy.std(speed)
print(x)
```

```python
## To calculate the variance you have to do as follows:
## Find the mean:
## (32+111+138+28+59+77+97) / 7 = 77.4
## For each value: find the difference from the mean:

```

```python
speed = [32,111,138,28,59,77,97]
x = np.mean(speed)
print(x)
```

```python
print('Variance:', np.var(speed))
var = 0
for i in speed:
    n = i - x
    sqval = n * n
    var += sqval

print("Variance:",var/7)
```
